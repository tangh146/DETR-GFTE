convert pubtabnet annotation format to GCN format for the GCN model

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))

import json
from utils import get_table_grid

def ptn2gcn(input_path, output_path, split="train"):
    with open(input_path, 'r', encoding="utf8") as infile, open(output_path, 'w', encoding="utf8") as outfile:
        max_horz = 0
        max_vert = 0
        impurities = 0

        count = 0
        
        for line in infile:
            # Parse the JSON line
            data = json.loads(line)
            
            if data["split"] == split:
                bbox = []
                bbox_indices = []
                for index, cell in enumerate(data['html']['cells']):
                    if 'bbox' in cell:
                        xmin, ymin, xmax, ymax = cell['bbox']
                        width = xmax - xmin
                        height = ymax - ymin
                        bbox.append([xmin, ymin, width, height])
                        bbox_indices.append(index)

                pure = True

                html = data['html']['structure']['tokens']
                html = ''.join(html)

                thead_grid, tbody_grid = get_table_grid(html)
                table_grid = thead_grid + tbody_grid

                base_width = len(table_grid[0])
                for row in table_grid:
                    if len(row) != base_width:
                        pure = False
                
                if pure:

                    # horz, vert = [], []

                    # # 0 1 2 3 4 5   <- len 6
                    # for row_idx, row in enumerate(table_grid):
                    #     for col_idx, item in enumerate(row):
                    #         for tgt_idx in range(col_idx, len(row)):
                    #             if item != row[tgt_idx]:
                    #                 horz.append([item, row[tgt_idx]])
                    #         for tgt_idx in range(row_idx, len(table_grid)):
                    #             if item != table_grid[tgt_idx][col_idx]:
                    #                 vert.append([item, table_grid[tgt_idx][col_idx]])
                    
                    # if len(horz) > max_horz: max_horz = len(horz)
                    # if len(vert) > max_vert: max_vert = len(vert)
                
                    output = {
                        'filename': data['filename'],
                        'image_id': data['imgid'],
                        'html': data['html']['structure']['tokens'],
                        'gt_bboxes': bbox,
                        'gt_bbox_indices': bbox_indices
                    }

                    outfile.write(json.dumps(output) + '\n')

                    count += 1
                else:
                    impurities += 1

        print(f"max horz (for padding): {max_horz}")
        print(f"max vert: {max_vert}")
        print(f"deformity count: {impurities}")

input_jsonl_path = r"C:\Users\remote desktop\Downloads\pubtabnet\PubTabNet_2.0.0.jsonl"
output_jsonl_path = r"C:\Users\remote desktop\Downloads\DETR-GFTE\datasets\gcn\stage3\ptn2gcn\val.jsonl"

ptn2gcn(input_jsonl_path, output_jsonl_path, split="val")
```
